## Single Example

```python
response = simp.simple_image_download
search_term = "poison ivy plant"
term_snake = search_term.replace(" ", "_")
download_path = Path(download_dir, search_term)
destination_path = Path(dataset_base_path, "test_example", term_snake)

response().download(search_term, 1)
shutil.rmtree(str(destination_path))
shutil.move(str(download_path), str(destination_path))
for file in os.listdir(destination_path):
    os.rename(Path(destination_path, file), Path(destination_path, file.replace(" ", "_")))
```

```python
for file in os.listdir(destination_path):
    print(file)
    display(Image(filename=str(Path(destination_path, file)), height=300))
```
