```python
%%time
async def fetch_json(url, session):
    async with session.get(url) as response:
        return(await response.json())
    

async def get_responses(vals):
    async with aiohttp.ClientSession("https://jsonplaceholder.typicode.com") as sess:
        tasks = [
            asyncio.ensure_future(fetch_json(f"/todos/{url}", sess))
            for url in vals
        ]
        results = await asyncio.gather(*tasks)
    return(results)

def print_responses(vals):
    loop = asyncio.get_event_loop()
    responses = loop.run_until_complete(get_responses(vals))
    # print(responses)
    return(responses)


start_time = time.time()
reses = print_responses([*range(1,200)])
for res in reses:
    print(res['title'])
print(time.time() - start_time)
```
